error='coerce'-> convert missing value invalid format to NaT->Not a time

```python

temp_data = pd.read_csv("/content/crime_dataset_india.csv")

temp_data['Time of Occurrence'] = pd.to_datetime(temp_data['Time of Occurrence'], format='%d-%m-%Y %H:%M', errors='coerce').dt.time

def get_day_night(time_obj):
    if pd.isnull(time_obj):
        return None
    hour = time_obj.hour
    if 6 <= hour < 18:
        return 'Day'
    else:
        return 'Night'

temp_data['day_night'] = temp_data['Time of Occurrence'].apply(get_day_night)
display(temp_data[['Time of Occurrence', 'day_night']].head(6))
```

```python
data.head(2)
```

```python
# covert date column to standard format, utc=true:same datetime across zones
cols=["Date Reported","Date of Occurrence","Time of Occurrence","Date Case Closed"]
data[cols]=data[cols].apply( lambda x: pd.to_datetime(x,format='%d-%m-%Y %H:%M',errors='coerce',dayfirst=True,utc=True))

data
```

```python
#fill missing weapon used values with unknown
data["Weapon Used"].fillna(value="Unknown")
```

```python
data.drop(columns="Report Number")
```

```python
data.drop(columns="Date of Occurrence",inplace=True)
```

```python
data
```

```python
data["Date Case Closed"]=data["Date Case Closed"].fillna("not closed").astype('object')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cases_by_city = data['City'].value_counts()

plt.figure(figsize=(8,5))
sns.barplot(x=cases_by_city.index, y=cases_by_city.values, palette='viridis')
plt.title('Number of Crime Cases by City')
plt.xlabel('City')
plt.ylabel('Number of Cases')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python

# 1. Crime Distribution by City (Top 10)
plt.subplot(3, 2, 1)
city_data = data['City'].value_counts().nlargest(10).reset_index()
city_data.columns = ['City', 'Count']
plt.barh(city_data['City'], city_data['Count'])
plt.title('Where Crimes Happen Most: City Hotspots', pad=30, fontsize=10)
plt.xlabel('Reported Cases', fontsize=5)
plt.ylabel('', fontsize=5)
```

```python
top_5_cities=data["City"].value_counts().head(5).index.tolist()
top_5_cities
```

```python
#top 7 city with max crime rate over 5 years
crime_by_city_year = data.groupby(['City', 'year']).size().reset_index(name='crime_count')
top_7_cities = data['City'].value_counts().head(7).index.tolist()
crime_top_7_cities = crime_by_city_year[crime_by_city_year['City'].isin(top_7_cities)]


plt.figure(figsize=(8,4))
sns.lineplot(data=crime_top_7_cities, x='year', y='crime_count', hue='City', marker='.')
plt.title('Crime Counts in Top 7 Cities by Year')
plt.xlabel('Year')
plt.ylabel('Number of Crimes (Count)')


plt.legend(title='City')

plt.show()
```

```python
#victim age
plt.figure(figsize=(8, 4))

plt.hist(data['Victim Age'],bins=100,edgecolor="green")
plt.title('Number of Crime Cases by Victim Age',pad=20)
plt.xlabel('Victim Age')
plt.ylabel('Number of Cases')
plt.show()
```

```python
var=data["Victim Age"].value_counts().sort_index()
```

```python
plt.figure(figsize=(8,4))
plt.plot(var.index,var.values)
plt.xlabel("Victim Age")
plt.ylabel("Crime Count")
plt.title("Crime Count by Victim Age")
plt.show()
```

```python
gen=data['Victim Gender'].value_counts()
plt.pie(gen.values,labels=gen.index,autopct="%1.1f%%")
plt.title("Gender affected by crime distribution")
gen
```

```python
#crime rate over year/months
plt.figure(figsize=(11,5))

year_month_data=data.groupby(["year","month"]).size().reset_index(name="total count")

sns.lineplot(data=year_month_data, x="month", y="total count",hue="year",marker="o")

plt.xlabel("Month")
plt.ylabel("Crime Count")
plt.title("Crime Count over year")
plt.xticks(range(1,13),['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec'])
plt.legend(title="Year")
plt.tight_layout()
plt.show()
```

```python
#distribution of case closing status by cities

city_data=data.groupby(['City',"Case Closed"]).size().unstack()

plt.figure(figsize=(11,5))
city_data.plot(kind='bar',stacked=True)
plt.xlabel("City")
plt.ylabel("Count of cases")
plt.title("Case Closed Status by City")
plt.legend(title="Case Closed")
plt.tight_layout()
plt.show()
```
